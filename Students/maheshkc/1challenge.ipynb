Consider the function defined below.
 * Plot the function using [matplotlib](https://matplotlib.org/)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def g(t): 
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))

t1 = np.arange(0.0, 2.0, 0.01)

plt.figure(1)
plt.plot(t1, g(t1), 'k')
plt.show()


def f(a,t):
    return 1- (np.exp(-1*a*t))



def L2norm(a): ## Function to find the L2 norm between g(t) and f(a,t) for a paticular value of a for all values of t
    
    var=0
    for t in (t1):
         var+=np.square(g(t)-f(a,t))
    return var

min_a = minimize_scalar(L2norm)

plt.plot(t1,g(t1),'k')
plt.plot(t1,f(min_a.x,t1),'m')
plt.show()

print ("Value of 'a' for minimum of L_2 norm between g(t) and f(t):  ", min_a.x)
print ("The mean square error: ", L2norm(min_a.x)/(len(t1)))
```
