Consider the class of functions of the form $f(t) = 1 - e^{-at}$ where $a$ is a free parameter.
 * Find the value of $a$ that minimizes the $L_2$ norm between $g(t)$ and $f(t)$ over the interval $[0, 2]$
 * Computer the residual mean squared error
 * Explain how you obtained $a$

```python
import numpy as np
import matplotlib.pyplot as plt


def g(t):
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))
def f(t,a):
    return 1-(np.exp(-a*t))


t2 = np.arange(0.0, 2.00, 0.01)
a2=np.arange(0.0,10.0,0.0001)
minN=float('inf')
mina=0
mse=0
rmse=0


for a in a2:
    N=(sum((abs(f(t2,a)-g(t2)))**2))**0.5
    if N<minN:  
        minN=N
        mina=a
        
mse=(minN*minN)/len(t2)
rmse=np.sqrt(mse)

#results
plt.figure(1)
plot1=plt.plot(t1, g(t2), 'k',label='g(t)')
plot2=plt.plot(t1, f(t2,mina), 'r',label='f(t)')
plt.legend(loc='best')
plt.title('challenge 1')
plt.xlabel('t')
plt.ylabel('function')
plt.show()
print("The valus of min a is ",mina)
print("The value of L2 norm between f(t) and g(t) ",minN)
print("The value of mean squared error is",mse)
print("The value of root mean squared error is",rmse)
```
