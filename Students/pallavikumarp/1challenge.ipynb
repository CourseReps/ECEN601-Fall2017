Consider the function defined below.
 * Plot the function using [matplotlib](https://matplotlib.org/)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def g(t): # function g(t)
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))

t1 = np.arange(0.0, 2.0, 0.01)

plt.figure(1)
plt.plot(t1, g(t1), 'b')
plt.show()


def f(a,t):# function f(a,t)
    return 1- (np.exp(-1*a*t))

#Solution

def Findnorm(a): ## Function to find the L2 norm between g(t) and f(a,t) for a paticular value of a for all values of t
	Temp_value=0			#Initialize
	for t in (t1):
		Temp_value+=np.square(g(t)-f(a,t))
	return Temp_value

Result = minimize_scalar(Findnorm)

print ("Value of a that minimizes the L_2 norm between g(t) and f(t):  ", Result.x)
print ("The mean square error (MSE): ", Findnorm(Result.x)/(len(t1)))

plt.plot(t1,g(t1),'b')
plt.plot(t1,f(Result.x,t1),'r')
plt.show()
```
