```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def g(t):
    return 1 - (np.exp(-2 * t) * np.cos(np.pi * t))

def f(a, t):
    return 1 - (np.exp(-a * t))

t1 = np.arange(0.0, 2.0, 0.001)

# Find the value of a and the mse
def minimum(a):
    L2 = 0
    for t in (t1):
        L2=(L2+(g(t)-f(a,t))**2)
        
    return L2**0.5

mina = minimize_scalar(minimum)
length=len(t1)
mse= np.square(minimum(mina.x))/length

print("the minimum a is  ",mina.x)
print("the resident mean square error is " ,mse)

#plot
plt.figure(1)
a1=np.arange(1,100,0.1)
plt.plot(a1,minimum(a1))
plt.xlabel("a")
plt.ylabel("L2")
plt.show()
```

```python
# how I get a
# firstly, I define the function f and g. Then according to the definition of L2 norm, I get the L2 norm by using a "for loop"
# by using the minimize_scalar, I get the minimum a and by calculating many L2 values at different a, I plot the diagram. 
```
