```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(t):
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))

t1 = np.arange(0.0, 2.0, 0.01)

plt.figure(1)
plt.plot(t1, g(t1), 'k')
plt.show()
```

```python
import scipy
from scipy import spatial
from scipy.spatial import distance
from scipy.spatial.distance import euclidean
```

```python
def f(t):
    return 1 - (np.exp(-a*t))
t1 = np.arange(0.0, 2.0, 0.01)
a=-10
u= g(t1)
v= f(t1)
d=scipy.spatial.distance.euclidean(u,v)
while a<=10:
    u= g(t1)
    v= f(t1)
    d1=scipy.spatial.distance.euclidean(u,v)
    d=min(d,d1)
    if d1==d:
        a_optimal=a
        
    a=a+0.001
```

```python
a=a_optimal
def f(t):
    return 1 - (np.exp(-a_optimal*t))

t1 = np.arange(0.0, 2.0, 0.01)
u= g(t1)
v= f(t1)
plt.figure(2)
plt.plot(t1, f(t1), 'b')
plt.show()
```
