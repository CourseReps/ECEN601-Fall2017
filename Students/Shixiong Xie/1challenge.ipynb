Consider the class of functions of the form $f(t) = 1 - e^{-at}$ where $a$ is a free parameter.
 * Find the value of $a$ that minimizes the $L_2$ norm between $g(t)$ and $f(t)$ over the interval $[0, 2]$
 * Computer the residual mean squared error
 * Explain how you obtained $a$

```python
import numpy as np
import matplotlib.pyplot as plt

##define g(t) and f(t)
def g(t):
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))
def f(t,a):
    return 1 - (np.exp(-a*t))

##define values
t1 = np.arange(0.0, 2.00, 0.01)
a1 = np.arange(0.0, 20.00, 0.0001)
Minl = float ('inf')
Mina = 0
MSE  = 0
##loop
for a in a1:
       L=(sum((abs(f(t1,a)-g(t1)))**2))**0.5
       if L<Minl:
        Minl = L
        Mina = a
        
MSE=(Minl*Minl)/len(t1)

#show the result
plt.figure(1)
plot1=plt.plot(t1, g(t1), 'k',label='g(t)')
plot2=plt.plot(t1, f(t1,Mina),'r',label='f(t)')
plt.title('challenge1')
plt.show()
print("The values of Mina is", Mina)
print("The values of L2 is", Minl)
print("The values of MSE is", MSE)
```
