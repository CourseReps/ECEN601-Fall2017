Consider the class of functions of the form $f(t) = 1 - e^{-at}$ where $a$ is a free parameter.
 * Find the value of $a$ that minimizes the $L_2$ norm between $g(t)$ and $f(t)$ over the interval $[0, 2]$
 * Computer the residual mean squared error
 * Explain how you obtained $a$

```python
import numpy as np
import matplotlib.pyplot as plt

## define the function g(t) and f(t)
def g(t):
    return 1 - (np.exp(-2*t) * np.cos(np.pi*t))
def f(t,a):
    return 1 - (np.exp(-a*t))
## define the range of t1
t1 = np.arange(0.0, 2.0, 0.01)
## define the range of a and decide the number of a
ara = np.arange(0.0, 20.0, 0.001) 
## define the minimum value of L2 form, because the number of result is undecided, so I used float('inf')for comparision bwtween the unbounded value
minimum_of_L2= float('inf')
## I defined the start value of minimum value of a to be 0 
minimum_of_a=0
rms = 0
## I used for loop for getting the minimum value of L2, and accoridng to the min value of L2 to get the min value of a
for a in ara:
    L2 = np.sum((g(t1)-f(t1,a))**2)**0.5
    if L2 < minimum_of_L2:
        minimum_of_L2 = L2
        minimum_of_a=a
        ## here is the value of rms which is L2^2/the length of t1
        rms = sum((g(t1)-f(t1,a))**2)/len(t1)
print("the value of L2 norm is", minimum_of_L2)
print("the value of a is",minimum_of_a)
print("the value of rms is",rms)
plt.figure(1)
plt.plot(t1, g(t1), 'k')
plt.figure(2)
plt.plot(t1,f(t1,minimum_of_a),'r')
plt.show() 
```
